```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# 1. 폰트 및 마이너스 기호 설정 (환경에 맞게 주석 해제)
plt.rc('font', family='Malgun Gothic') # Windows
# plt.rc('font', family='AppleGothic') # Mac
plt.rcParams['axes.unicode_minus'] = False 

# 2. 통계적 유의미성 계산
# [산점도 지표] 피어슨 상관계수(r)와 결정계수(R-squared) 계산
correlation = df['전용면적'].corr(df['거래금액'])
r_squared = (correlation ** 2) * 100  # 백분율(%) 변환

# [히스토그램 지표] 왜도(Skewness, 데이터가 한쪽으로 쏠린 정도) 계산
skewness = df['거래금액'].skew()
skew_text = "정규분포에 가까움" if abs(skewness) < 1 else "한쪽으로 심하게 치우침"

# 3. 도화지 준비 (텍스트가 들어갈 하단 여백을 위해 figsize 조정)
fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# --- 첫 번째 그래프: 히스토그램 ---
sns.histplot(data=df, x='거래금액', bins=50, kde=True, ax=axes[0], color='cornflowerblue')
axes[0].set_title(f'거래금액 분포\n(분포 상태: {skew_text})', fontsize=14, fontweight='bold')
axes[0].set_xlabel('거래금액')
axes[0].set_ylabel('빈도수')

# --- 두 번째 그래프: 산점도 ---
sns.scatterplot(data=df, x='전용면적', y='거래금액', alpha=0.5, ax=axes[1], color='tomato')
# 회귀선(추세선)을 추가하여 관계를 더 명확히 보여줌
sns.regplot(data=df, x='전용면적', y='거래금액', scatter=False, ax=axes[1], color='darkred', line_kws={'linestyle':'--'})
axes[1].set_title(f'전용면적 vs 거래금액\n(설명력: 두 변수의 관계성 {r_squared:.1f}%)', fontsize=14, fontweight='bold')
axes[1].set_xlabel('전용면적')
axes[1].set_ylabel('거래금액')

# 4. 분석 결과 종합 코멘트를 하단에 출력
insight_text = (
    f"💡 [종합 분석 결과]\n"
    f"1. 히스토그램: 거래금액 데이터의 왜도는 {skewness:.2f}로, {skew_text} 상태를 보여줍니다.\n"
    f"2. 산점도: 전용면적이 거래금액의 변동성을 약 {r_squared:.1f}% 설명하고 있습니다.\n"
    f"▶ 결론: 이 데이터에서는 전용면적과 거래금액의 뚜렷한 관계성을 보여주는 '산점도' 분석이 매우 유의미하며 적합도가 높습니다."
)

plt.figtext(0.5, 0.02, insight_text, ha="center", fontsize=13, 
            bbox={"facecolor":"lightyellow", "alpha":0.8, "pad":10, "boxstyle":"round,pad=0.5"})

# 5. 그래프 레이아웃 조정 및 출력
plt.tight_layout(rect=[0, 0.15, 1, 1]) # 하단 텍스트 박스를 위한 여백 확보
plt.show()
```

```python
# 필요 라이브러리 설치: pip install streamlit pandas plotly
import streamlit as st
import pandas as pd
import numpy as np
import plotly.express as px

# 1. 페이지 기본 설정 (가장 상단에 위치해야 함)
st.set_page_config(page_title="부동산 인터랙티브 보고서", page_icon="🏢", layout="wide")

# 2. 샘플 데이터 생성 (실제 df가 있다면 이 부분을 대체하세요)
@st.cache_data
def load_data():
    np.random.seed(42)
    areas = ['A동', 'B동', 'C동', 'D동', 'E동']
    data = {
        '법정동': np.random.choice(areas, 500),
        '전용면적': np.random.uniform(30, 150, 500),
        '건축년도': np.random.randint(1990, 2023, 500)
    }
    df = pd.DataFrame(data)
    # 전용면적과 동별 프리미엄에 비례하여 거래금액 생성
    area_premium = {'A동': 1.5, 'B동': 1.2, 'C동': 1.0, 'D동': 0.8, 'E동': 0.6}
    df['거래금액'] = df['전용면적'] * df['법정동'].map(area_premium) * np.random.uniform(0.8, 1.2, 500) * 1000
    df['거래금액'] = df['거래금액'].round(0)
    return df

df = load_data()

# 3. 사이드바 구성 (PPT의 슬라이드 목차 역할)
st.sidebar.title("📑 목차 네비게이션")
pages = [
    "1. 데이터 수집 개요", 
    "2. 법정동별 평균 거래금액", 
    "3. 거래금액 분포 (히스토그램)", 
    "4. 면적-금액 상관관계 (산점도)",
    "5. 결론 및 제언"
]
selected_page = st.sidebar.radio("슬라이드를 선택하세요:", pages)

st.sidebar.markdown("---")
st.sidebar.info("💡 **Tip:** 좌측 메뉴를 클릭하여 PPT 슬라이드를 넘기듯 내용을 확인할 수 있습니다.")

# 4. 페이지별 내용 렌더링
if selected_page == "1. 데이터 수집 개요":
    st.title("🏢 1. 데이터 수집 개요")
    st.markdown("""
    * **수집 대상:** 주요 법정동 부동산 실거래가 데이터
    * **주요 변수:** 법정동, 거래금액, 전용면적, 건축년도 등
    * **분석 목적:** 지역별 시세 파악 및 가격 결정 요인 도출
    """)
    # 인터랙티브 요소: 원본 데이터 프레임 토글 보기
    if st.checkbox("원본 데이터 미리보기"):
        st.dataframe(df.head(15), use_container_width=True)

elif selected_page == "2. 법정동별 평균 거래금액":
    st.title("📊 2. 법정동별 평균 거래금액")
    
    # 인터랙티브 요소: 건축년도 슬라이더 필터 적용
    min_year, max_year = int(df['건축년도'].min()), int(df['건축년도'].max())
    selected_years = st.slider("건축년도 필터", min_value=min_year, max_value=max_year, value=(2000, 2020))
    
    filtered_df = df[(df['건축년도'] >= selected_years[0]) & (df['건축년도'] <= selected_years[1])]
    
    df_mean = filtered_df.groupby('법정동')['거래금액'].mean().reset_index()
    df_mean = df_mean.sort_values(by='거래금액', ascending=False)
    
    # Plotly를 이용한 인터랙티브 막대그래프 (마우스 오버, 확대 지원)
    fig = px.bar(df_mean, x='법정동', y='거래금액', color='법정동', title=f"{selected_years[0]}~{selected_years[1]}년 건축 매물 평균 거래금액")
    st.plotly_chart(fig, use_container_width=True)

elif selected_page == "3. 거래금액 분포 (히스토그램)":
    st.title("📈 3. 거래금액 분포 분석")
    
    # 인터랙티브 요소: 특정 동 선택 드롭다운
    selected_dong = st.selectbox("분석할 법정동을 선택하세요:", ["전체"] + list(df['법정동'].unique()))
    
    plot_df = df if selected_dong == "전체" else df[df['법정동'] == selected_dong]
    
    # 인터랙티브 요소: Bin(막대) 개수 조절 슬라이더
    bins = st.slider("히스토그램 막대(Bin) 개수 조절", min_value=10, max_value=100, value=30, step=10)
    
    fig = px.histogram(plot_df, x='거래금액', nbins=bins, title=f"[{selected_dong}] 거래금액 분포", color_discrete_sequence=['cornflowerblue'])
    st.plotly_chart(fig, use_container_width=True)

elif selected_page == "4. 면적-금액 상관관계 (산점도)":
    st.title("🔗 4. 전용면적과 거래금액의 관계")
    st.markdown("마우스를 점 위에 올려 매물의 세부 정보를 확인해보세요.")
    
    # Plotly 인터랙티브 산점도 (hover_data로 마우스 오버 시 추가 정보 제공)
    fig = px.scatter(df, x='전용면적', y='거래금액', color='법정동', opacity=0.7, 
                     hover_data=['건축년도'], title="면적 vs 금액 산점도")
    st.plotly_chart(fig, use_container_width=True)

elif selected_page == "5. 결론 및 제언":
    st.title("💡 5. 결론 및 제언")
    st.success("""
    **분석 결과 요약**
    1. **지역별 격차:** 법정동에 따라 평균 거래금액의 뚜렷한 프리미엄이 관찰됩니다. (필터링 대시보드 참고)
    2. **면적의 중요성:** 전용면적은 거래금액과 가장 강한 선형 관계를 보여줍니다.
    
    **향후 과제**
    * 추가적인 외부 요인(금리, 학군 데이터 등)을 결합한 다차원 분석 모델 고도화 필요
    """)
    st.balloons() # 인터랙티브 재미 요소 (결론 페이지 진입 시 풍선 애니메이션)
```

```python
from deepface import DeepFace

def check_similarity(img1_path, img2_path):
    try:
        print("사진을 분석 중입니다. 잠시만 기다려주세요...")
        
        # DeepFace.verify() 함수로 두 이미지 비교
        # model_name으로 'VGG-Face', 'Facenet', 'ArcFace' 등을 선택할 수 있습니다.
        result = DeepFace.verify(img1_path=img1_path, 
                                 img2_path=img2_path, 
                                 model_name='Facenet')
        
        # 결과 분석
        is_same_person = result['verified'] # 동일인 여부 (True/False)
        distance = result['distance']       # 두 얼굴 간의 거리 (낮을수록 닮음)
        
        # 거리(distance) 값을 직관적인 닮은 비율(%)로 간단히 변환 (Facenet 기준 임계값 약 0.4 적용)
        # 완벽한 비례식은 아니지만, 직관적인 이해를 돕기 위한 예시 계산식입니다.
        similarity_score = max(0, (1 - (distance / 0.8))) * 100
        
        print("\n--- 분석 결과 ---")
        if is_same_person:
            print(f"✅ 분석 결과: 두 사진은 '동일인'으로 판단됩니다!")
        else:
            print(f"❌ 분석 결과: 두 사진은 '다른 사람'으로 판단됩니다.")
            
        print(f"📊 닮은 정도 (유사도): 약 {similarity_score:.1f}%")
        print(f"   (참고 - 얼굴 간 거리: {distance:.4f} / 낮을수록 가까운 사이)")
        
    except ValueError as e:
        print(f"\n[오류 발생] 사진에서 얼굴을 찾을 수 없습니다. ({e})")
        print("얼굴이 정면을 향하고 선명하게 나온 사진을 사용해 주세요.")

# 테스트 실행 (본인의 사진 파일 이름으로 변경하세요)
# 예: check_similarity('my_face1.jpg', 'my_face2.jpg')
image1 = "photo1.jpg" 
image2 = "photo2.jpg"

check_similarity(image1, image2)
```
