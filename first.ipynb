```python
num = 1
total_sum = 0

# 조건이 True이므로 무한 루프를 돌되, 내부에서 조건이 만족되면 break로 탈출합니다.
while True:
    total_sum += num  # 숫자를 누적해서 더합니다.
    
    # 합이 20이 넘는지 확인합니다.
    if total_sum > 20:
        # 조건이 충족되면 원하는 값을 출력하고 반복문을 종료합니다.
        print(f"정수: {num}, 합: {total_sum}")
        break
```

```python
import random

# 컴퓨터가 1부터 100 사이의 정답 숫자를 하나 무작위로 뽑습니다.
com = random.randint(1, 100)

print("컴퓨터가 1부터 100 사이의 숫자를 생각했습니다. 맞춰보세요!")

while True:
    # 사용자에게 숫자를 입력받습니다.
    user_input = int(input("숫자를 입력하세요: "))
    
    # 사용자가 입력한 값과 컴퓨터의 정답(com)을 비교합니다.
    if user_input == com:
        print("정답입니다! 게임을 종료합니다.")
        break  # 정답을 맞췄으므로 while 루프를 탈출합니다.
        
    elif user_input < com:
        print("작다! (더 큰 숫자를 입력해보세요.)")
        
    elif user_input > com:
        print("크다! (더 작은 숫자를 입력해보세요.)")
```

```python
# 1. 어디까지 진행할지 입력받아 숫자로 저장 (예: 20)
target = int(input('1부터 얼마까지 진행할까요?(예:20) : '))

# 2. 시작할 숫자 설정
current_num = 1

# 3. 시작 숫자가 목표 숫자에 도달할 때까지 반복
while current_num <= target:
    
    # 숫자를 문자열(글자)로 바꾸어 '3', '6', '9'가 포함되어 있는지 확인
    # 이렇게 하면 3, 6, 9뿐만 아니라 13, 16, 19 등도 모두 '짝' 처리가 됩니다!
    if '3' in str(current_num) or '6' in str(current_num) or '9' in str(current_num):
        print('짝', end = " ")
    else:
        print(current_num, end = " ")
    
    # 다음 숫자로 넘어가기
    current_num += 1
```

```python
num1 = 1
num2 = 15

while num1 <= num2:
    # 전체 너비에서 별이 차지하는 만큼 공간을 계산해 왼쪽에 공백을 넣습니다.
    # (num2 - num1) // 2 만큼 공백을 출력해 줄을 맞춥니다.
    space = (num2 - num1) // 2
    print(" " * space + "*" * num1)
    
    num1 += 2
```

```python
import random

# 컴퓨터가 사용할 단어 사전 (원하는 단어를 더 추가할 수 있어요!)
word_dict = ["가나다", "다람쥐", "쥐돌이", "이발소", "소나무", "무지개", "개구리", "리본", "본드", "드라마", "마술"]

print('★ 컴퓨터와 함께하는 끝말잇기 게임 ★')
print('종료하고 싶다면 "종료"를 입력하세요.')

# 첫 제시어 무작위 선정
current_word = random.choice(word_dict)
print(f'\n[컴퓨터 제시어]: {current_word}')

while True:
    # 플레이어 입력
    user_word = input('[당신]: ').strip()
    
    # 종료 조건
    if user_word == '종료':
        print('게임이 종료되었습니다.')
        break
        
    # 글자 수 제한 (최소 2글자)
    if len(user_word) < 2:
        print('❌ 2글자 이상의 단어를 입력해 주세요.')
        continue
        
    # 끝말잇기 규칙 확인 (이전 단어의 끝 글자 == 내 단어의 첫 글자)
    if current_word[-1] != user_word[0]:
        print(f'❌ 틀렸습니다! "{current_word[-1]}"(으)로 시작하는 단어여야 합니다. 당신의 패배!')
        break
        
    # 플레이어가 통과했을 때, 컴퓨터의 차례
    print('-> 통과!')
    
    # 컴퓨터가 이어받을 수 있는 단어 찾기
    possible_words = [w for w in word_dict if w[0] == user_word[-1]]
    
    if not possible_words:
        print(f'\n[컴퓨터]: 으악, "{user_word[-1]}"(으)로 시작하는 단어가 없어! 당신의 승리! 🎉')
        break
        
    # 컴퓨터가 찾은 단어 중 하나를 무작위로 선택
    current_word = random.choice(possible_words)
    print(f'\n[컴퓨터]: {current_word}')
```

```python
import random

print("=== 가위바위보 게임을 시작합니다! ===")

choices = ["가위", "바위", "보"]

# 🔴 승, 무, 패를 저장할 변수를 0으로 초기화합니다.
win = 0
draw = 0
lose = 0

while True:
    player = input("가위, 바위, 보 중 하나를 입력하세요 (종료하려면 '종료' 입력): ")

    if player == "종료":
        print("\n=== 최종 전적 ===")
        print(f"🏆 {win}승 | 🤝 {draw}무 | 😢 {lose}패")
        print("게임을 종료합니다. 다음에 또 봐요!")
        break

    if player not in choices:
        print("⚠️ 잘못된 입력입니다! '가위', '바위', '보' 중에서만 입력해주세요.\n")
        continue

    computer = random.choice(choices)
    print(f"나: {player} vs 컴퓨터: {computer}")

    # 승패 판정 및 전적 카운트
    if player == computer:
        print("🤝 비겼습니다!")
        draw += 1  # 무승부 1 증가 (draw = draw + 1 과 같아요)
    elif (
        (player == "가위" and computer == "보")
        or (player == "바위" and computer == "가위")
        or (player == "보" and computer == "바위")
    ):
        print("🎉 이겼습니다!")
        win += 1  # 승리 1 증가
    else:
        print("😢 졌습니다...")
        lose += 1  # 패배 1 증가

    # 📢 매 판이 끝날 때마다 현재 전적을 보여줍니다.
    print(f"현재 전적: {win}승 {draw}무 {lose}패")
    print("-" * 30)
```

```python
def min_num(numbers):
    # 1. 일단 리스트의 맨 첫 번째 숫자(52)를 가장 작다고 가정하고 '기준값'으로 잡습니다.
    minimum = numbers[0]
    
    # 2. 리스트의 숫자들을 하나씩 꺼내며 기준값과 비교합니다.
    for i in numbers:
        # 만약 이번에 꺼낸 숫자(i)가 내가 알던 기준값(minimum)보다 더 작다면?
        if i < minimum:
            # 새로운 최솟값으로 임시 저장소를 업데이트합니다!
            minimum = i
            
    return minimum

# 테스트
print(min_num([52, 273, 32, 103, 275, 24, 57]))  # 출력: 24
```

```python
def isLeapYear(year):
    # 1. 400으로 나누어 떨어지면 무조건 윤년! (가장 강력한 조건)
    if year % 400 == 0:
        return True
    
    # 2. 위 조건은 통과 못 했지만, 100으로 나누어 떨어지면 평년!
    elif year % 100 == 0:
        return False
    
    # 3. 위 조건들 다 통과 못 했지만, 4로 나누어 떨어지면 윤년!
    elif year % 4 == 0:
        return True
    
    # 4. 그 외의 모든 해는 평년!
    else:
        return False

# --- 테스트 코드 ---
print(f"2024년: {isLeapYear(2024)}")  # True (4의 배수라 윤년)
print(f"1900년: {isLeapYear(1900)}")  # False (4의 배수지만 100의 배수라 평년)
print(f"2000년: {isLeapYear(2000)}")  # True (100의 배수지만 400의 배수라 윤년)
print(f"2026년: {isLeapYear(2026)}")  # False (평년)
```

```python
def add_mul(choice, *args):
    # 1. 만약 첫 번째 인자가 'add'라면 더하기를 수행합니다.
    if choice == 'add':
        result = 0
        for i in args:
            result += i
        return result
        
    # 2. 만약 첫 번째 인자가 'mul'이라면 곱하기를 수행합니다.
    elif choice == 'mul':
        result = 1  # 곱셈의 기준값은 1로 시작해야 합니다!
        for i in args:
            result *= i
        return result

# --- 테스트 코드 ---
result = add_mul('add', 1, 2, 3, 4, 5)
print(result)  # 출력: 15 (1+2+3+4+5)

result = add_mul('mul', 1, 2, 3, 4, 5)
print(result)  # 출력: 120 (1*2*3*4*5)
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

# 1. 대상 URL 설정
url = "https://finance.naver.com/news/mainnews.naver"

# 네이버 서버가 로봇으로 인식해 차단하는 것을 방지하기 위한 헤더 설정
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

# 2. 페이지 HTML 가져오기
response = requests.get(url, headers=headers)
response.encoding = 'euc-kr'  # 네이버 금융 한글 깨짐 방지

soup = BeautifulSoup(response.text, "html.parser")

# 3. 주요 뉴스 기사 블록(들) 전체 찾기 (결과는 리스트 형태)
articles = soup.select('.block1')

print(f"--- 총 {len(articles)}개의 주요 뉴스를 찾았습니다 ---\n")

# 4. 반복문(for)을 돌며 각 기사 안에서 원하는 정보만 쏙쏙 추출하기
for article in articles:
    # 1) 기사 제목 및 링크 (제목은 .articleSubject > a 안에 들어있음)
    title_tag = article.select_one('.articleSubject > a')
    
    if title_tag:
        title = title_tag.text.strip()
        # 상대 경로로 되어있는 링크를 절대 경로(https://...)로 변환
        link = urljoin(url, title_tag['href'])
    else:
        continue # 제목이 없는 비정상 항목은 패스
        
    # 2) 기사 요약 (.articleSummary)
    # 요약문 뒤에 붙는 신문사, 날짜 등의 텍스트를 제외하고 순수 요약 내용만 추출하기 위해 text 대신 내용 정리
    summary_tag = article.select_one('.articleSummary')
    summary = ""
    if summary_tag:
        # 텍스트 추출 후 공백 제거
        full_summary_text = summary_tag.text.strip()
        # 네이버 특성상 요약, 신문사, 날짜가 한 태그 안에 뭉쳐있는 경우가 많아 줄바꿈 등으로 분리 가공이 필요할 수 있습니다.
        # 가장 깔끔하게 요약만 분리하기 위해 텍스트의 앞부분만 가져옵니다.
        summary = full_summary_text.split('\n')[0].strip()

    # 3) 신문사 (.press)
    press_tag = article.select_one('.press')
    press = press_tag.text.strip() if press_tag else "정보 없음"

    # 4) 작성 날짜 (.wdate)
    wdate_tag = article.select_one('.wdate')
    wdate = wdate_tag.text.strip() if wdate_tag else "정보 없음"

    # 5) 출력 결과 확인
    print(f"📰 기사제목: {title}")
    print(f"📝 기사요약: {summary}")
    print(f"🏢 신문사  : {press}")
    print(f"📅 작성날짜: {wdate}")
    print(f"🔗 링크    : {link}")
    print("-" * 50)
```
